<h2 style="color:#6a3093; margin-top:0;">3. Limpieza de datos</h2>
<h3 style="color:#8e5cc7;">3.1 Manejo de valores faltantes</h3>
<p>Se detectaron valores nulos en <code>promedio_calificaciones</code> y 
<code>puntaje_productividad</code>. Antes de decidir cómo tratarlos, se 
comparó la media y la mediana de ambas columnas para evaluar si existen 
outliers que sesguen los datos.</p>
</div>

```python
df['promedio_calificaciones'].mean()
```

```python
df['promedio_calificaciones'].median()
```

```python
df['puntaje_productividad'].mean()
```

```python
df['puntaje_productividad'].median()
```

<div style="background-color:#eaf7gc; border-left: 5px solid #4CAF50; padding: 15px; border-radius: 5px; margin-top:10px;">
<b>✅ Hallazgo:</b> La media y la mediana son similares en ambas columnas, 
lo que indica que no hay outliers fuertes distorsionando la distribución. 
Por lo tanto, se usará <code>mediana (.median())</code> para imputar los valores faltantes.
</div>

```python
mediana_calificaciones = df['promedio_calificaciones'].median()
df['promedio_calificaciones'] = df['promedio_calificaciones'].fillna(mediana_calificaciones)
```

```python
mediana_productividad = df['puntaje_productividad'].median()
df['puntaje_productividad'] = df['puntaje_productividad'].fillna(mediana_productividad)
```
